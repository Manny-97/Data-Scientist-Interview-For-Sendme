```python
#Importing all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from datetime import datetime
```

```python
#Reading the raw data using pandas
raw = pd.read_excel(r"C:\Users\USER\Downloads\Raw Data.xlsx")
index_unnamed = raw.columns.get_loc('Unnamed: 0')
print(index_unnamed)
print(raw.head())
```

```python
#Using regex to extract strings
date_reg = r'(?<!\d)\d{10}(?!\d)'
number = []
for value in raw['Unnamed: 0'].values:
    if type(value) == float:
        number.append(value)
```

```python
#Storing the strings into a list
number = [x for x in number if x is not np.nan]
```

```python
number
```

```python
#Formatting the phone numbers properly
clean_num = []
for value in number:
    if len(str(value)) == 12:
        value = float('234'+str(value))
    else:
        value = value
    clean_num.append(value)
    
```

```python
#Generating the Index of the phone numbers in the raw data
index_num = []
for val in number:
    a = np.where(raw['Unnamed: 0']==val)
    index_num.append(a[0][0])
    if len(a[0])==2:
        index_num.append(a[0][1])
    elif len(a[0]) == 3:
        index_num.append(a[0][1])
        index_num.append(a[0][2])
    print(a[0])
index_num = sorted(list(set(index_num)))
print(index_num)
```

```python
data['Contact'] = clean_num

```

```python
#Converting the Contact column to numeric data type
data['Contact'] = pd.to_numeric(data['Contact'], downcast='integer')
```

```python
#Using regex to extract date
pattern = r'([0-9]{2}\/[0-9]{2}\/[0-9]{4})'
date_col = []
for row in range(0, len(raw)):
    date = re.search(pattern, str(raw.iat[row, index_unnamed]))
    date_col.append(date)
```
